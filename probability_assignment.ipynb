Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

    -->School population:
	•	60% boys
	•	40% girls

You want equal representation of both genders in the sample.

Which method should you use?
Stratified Sampling
Why?

Because in stratified sampling, the population is divided into groups (strata) like boys and girls, and you can pick equal numbers from each group even if their actual proportions are different in the population.

Simple random sampling would not guarantee equal representation.


Question 5: The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm.

Find the sampling error

    -->Population average height = 160 cm
Sample average height = 158 cm

Sampling Error

Sampling error = Population mean − Sample mean
 Sampling Error = 160 - 158 = 2 cm

The sampling error is 2 cm.

Question 9: A factory produces bulbs with 2% defective rate. If 5 bulbs are chosen at random, what is the
probability that all are non-defective?

    -->Defective rate = 2% → probability non-defective = 1-0.02=0.98. For 5 independent bulbs, probability all non-defective:
P=(0.98)^5.
Compute stepwise:
0.98^2=0.9604,
0.98^3=0.9604 x 0.98=0.941192,
0.98^4=0.941192 x 98=0.92236816,
0.98^5=0.92236816 x 98=0.9039207968 = 0.9039.

Answer: 0.98^5 = 0.9039208 (≈90.39%)


Question 10: Differentiate between discrete and continuous random variables with examples

   
    -->Discrete vs Continuous random variables
	•	Discrete random variable: takes countable, separate values (often integers).
	•	Example: number of defective bulbs in a box (0,1,2,…).
	•	Characterized by a probability mass function (PMF) giving P(X=x) for each possible x.
	•	Continuous random variable: takes values from an interval (uncountably many), can take any real number within a range.
	•	Example: time (in hours) until a machine fails, or the exact height of a person.
	•	Characterized by a probability density function (PDF); probabilities for exact values are 0, but probability over an interval [a,b] is fab f(x)\,dx.
